###### Libraries

```python
import pandas as pd
import json
import requests
```

###### Load

```python
# Load JSON data from the URL
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/module%201/Accessing%20Data%20Using%20APIs/jobs.json"
response = requests.get(url)
data = response.json()
```

```python
# Convert to DataFrame
df = pd.DataFrame(data)
```

```python
# Function to count jobs requiring Python
def count_python_jobs(df):
    return df['Key Skills'].str.contains("Python", case=False, na=False).sum()
```

```python
# Call the function
python_job_count = count_python_jobs(df)
print("Number of jobs requiring Python:", python_job_count)
```

```python
api_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/module%201/Accessing%20Data%20Using%20APIs/jobs.json"
```

```python
def get_number_of_jobs_T(technology):
    # Get the data from the API
    response = requests.get(api_url)
    data = response.json()

    # Convert to DataFrame
    df = pd.DataFrame(data)

    # Count jobs where 'Key Skills' includes the given technology
    number_of_jobs = df['Key Skills'].str.contains(technology, case=False, na=False).sum()

    return technology, number_of_jobs
```

```python
# Example call
print(get_number_of_jobs_T("Python"))
```

##### Excel

```python
# Import libraries required to create excel spreadsheet
import requests
import pandas as pd
from openpyxl import Workbook
```

```python
# Create a python list of all technologies for which you need to find the number of jobs postings
technologies = [
    "C", "C#", "C++", "Java", "JavaScript", "Python", "Scala",
    "Oracle", "SQL Server", "MySQL Server", "PostgreSQL", "MongoDB"
]
```

```python
# Create a workbook and select the active worksheet
wb = Workbook()
ws = wb.active
ws.title = "Job Postings"
```

```python
# Find the number of jobs postings for each of the technologies in the list.
# Write the technology name and the number of jobs postings into the excel spreadsheet.

# Write headers
ws.append(["Technology", "Number of Job Postings"])

# Count and write results
for tech in technologies:
    count = df['Key Skills'].str.contains(tech, case=False, na=False).sum()
    ws.append([tech, count])
```

```python
# Save into an excel spreadsheet named job-postings.xlsx
wb.save("job-postings.xlsx")
```

```python
import requests
import pandas as pd
from openpyxl import Workbook

# === 1. Make a POST request to httpbin.org ===
post_url = "http://httpbin.org/post"
post_payload = {"technology": "Python", "location": "Los Angeles"}
r_post = requests.post(post_url, data=post_payload)

# === 2. Make a GET request for comparison ===
r_get = requests.get("http://httpbin.org/get")

# === 3. Print and analyze both URLs ===
print("GET URL:", r_get.url)
print("POST URL:", r_post.url)

# === 4. Print and analyze request bodies ===
print("POST Request Body:", r_post.request.body)
print("POST Returned Form Data:", r_post.json()['form'])

# === 5. Load the Jobs API (simulating Jobs_API notebook) ===
jobs_api_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/module%201/Accessing%20Data%20Using%20APIs/jobs.json"
response = requests.get(jobs_api_url)
data = response.json()
df = pd.DataFrame(data)

# === 6. Function to find Python jobs in a specific location ===
def get_python_jobs_in_location(df, location):
    return df[(df['Key Skills'].str.contains("Python", case=False, na=False)) &
              (df['Location'].str.contains(location, case=False, na=False))].shape[0]

python_la_count = get_python_jobs_in_location(df, "Los Angeles")
print("Python jobs in Los Angeles:", python_la_count)

# === 7. Technologies to search ===
technologies = [
    "C", "C#", "C++", "Java", "JavaScript", "Python", "Scala",
    "Oracle", "SQL Server", "MySQL Server", "PostgreSQL", "MongoDB"
]

# === 8. Create workbook ===
wb = Workbook()
ws = wb.active
ws.title = "Job Postings"
ws.append(["Technology", "Number of Job Postings"])

# === 9. Count jobs and write to Excel ===
for tech in technologies:
    count = df['Key Skills'].str.contains(tech, case=False, na=False).sum()
    ws.append([tech, count])

# === 10. Save Excel file ===
wb.save("job-postings.xlsx")
print("Excel file 'job-postings.xlsx' saved successfully.")
```
